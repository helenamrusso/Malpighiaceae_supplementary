1) Import tables POS and NEG (Featurecount1000);

2) Convert them to binary matrices (0 and 1 - absence and presence);

3) Check how much time in each genera there are zeros and ones;

4) Check which (presence or absence) was majoritary and build a new column for the genera and the binary character;
    (10 times absent, 3 times present = absent
    3 times absent, 10 times present = present
    3 times present, 3 times absent = present)
    
5) At this point, we should have a matrix with genera in columns, classes in rows, and populated with 0 and 1;

6) The two matrices (POS and NEG) must then be compared, and if absent in one ionization mode and present in another,
    keep 1. Therefore: 
    0+0 = 0
    0+1=1
    1+1=1
    


```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
# Import files
df_pos = pd.read_csv('/Users/helenarusso/Documents/UNESP/Doutorado/UCSD/Malpighiaceae_AllSamples/Qemistree.nosync/Files_for_running_Qemistree/Subsequent analyses/Análises Rafael/Tabelas filogenia/Filogenia ok/featuretable_updated_class_POS_cutoff1000.tsv', sep='\t', index_col=[0])
df_neg = pd.read_csv('/Users/helenarusso/Documents/UNESP/Doutorado/UCSD/Malpighiaceae_AllSamples/Qemistree.nosync/Files_for_running_Qemistree/Subsequent analyses/Análises Rafael/Tabelas filogenia/Filogenia ok/featuretable_updated_class_NEG_cutoff1000.tsv', sep='\t', index_col=[0])

df_pos
```

```python
# Convert imported files to binary matrices
df_pos[df_pos>1] = 1
df_pos = df_pos
df_pos
```

```python
# Keep only genera name
df_pos.columns = [col_name.split('_')[0] for col_name in df_pos.columns]
df_pos
```

```python
#Make a list with column names
colnames = df_pos.columns.unique()
colnames
```

```python
# Here we separate wach genera for its presence (_1) or absence (_0). Those genera that present only one sample
#     (one column in the original table), are maintained as only one column (without sulfix)
# colunas de gêneros que aparecem apenas uma vez foram copiados tais quais do dataframe binário (bloco except)
df = pd.DataFrame()
for colname in colnames:
    try:
        df[colname + '_0'] = df_pos[colname].apply(pd.Series.value_counts, axis=1).fillna(0)[0]
        df[colname + '_1'] = df_pos[colname].apply(pd.Series.value_counts, axis=1).fillna(0)[1]
    except TypeError:
         df[colname] = df_pos[colname]
        
        
df_pos = df
df_pos
```

```python
#reatribute colnames - update colnames

colnames_bin = df_pos.columns.unique()
colnames_bin
```

```python
#create loop to compare columns with _0 and _1 sulfix. If _1 >= _0, equals 1 (presence)

df = pd.DataFrame()
for colname in colnames:
    if colname + '_0' in colnames_bin:
        col0 = colname + '_0'
        col1 = colname + '_1'
        df[colname] = np.where((df_pos[col1] >= df_pos[col0]), 1, 0)

df = df.set_index(df_pos.index)
df
```

```python
for colname in colnames:
    if colname + '_0' not in colnames_bin:
        df[colname] = df_pos[colname]

df_pos = df
df_pos
```

```python
#####Now we do the same for the negative (all in on cell now)

# Convert imported files to binary matrices
df_neg[df_neg>1] = 1
df_neg = df_neg
df_neg.head()

# Keep only genera name
df_neg.columns = [col_name.split('_')[0] for col_name in df_neg.columns]
df_neg

#Make a list with column names
colnames = df_neg.columns.unique()
colnames

# Here we separate wach genera for its presence (_1) or absence (_0). Those genera that present only one sample
#     (one column in the original table), are maintained as only one column (without sulfix)
# colunas de gêneros que aparecem apenas uma vez foram copiados tais quais do dataframe binário (bloco except)
df = pd.DataFrame()
for colname in colnames:
    try:
        df[colname + '_0'] = df_neg[colname].apply(pd.Series.value_counts, axis=1).fillna(0)[0]
        df[colname + '_1'] = df_neg[colname].apply(pd.Series.value_counts, axis=1).fillna(0)[1]
    except TypeError:
         df[colname] = df_neg[colname]
        
        
df_neg = df
df_neg


#reatribute colnames - update colnames

colnames_bin = df_neg.columns.unique()
colnames_bin


#create loop to compare columns with _0 and _1 sulfix. If _1 >= _0, equals 1 (presence)

df = pd.DataFrame()
for colname in colnames:
    if colname + '_0' in colnames_bin:
        col0 = colname + '_0'
        col1 = colname + '_1'
        df[colname] = np.where((df_neg[col1] >= df_neg[col0]), 1, 0)

df = df.set_index(df_neg.index)
df

#
for colname in colnames:
    if colname + '_0' not in colnames_bin:
        df[colname] = df_neg[colname]

df_neg = df
```

# Now we join the two dataframes (df_neg and df_pos)

```python
joined_posneg = df_pos.join(df_neg, how='outer', lsuffix='_pos', rsuffix='_neg').fillna('X')
# joined_posneg
```

```python
# Creating a list mapping the row numbers and origin (pos or neg) based on the the cells filled with 'X' and the 
# column name
mylist = [(x, joined_posneg.columns[y].split('_')[1]) for x, y in zip(*np.where(joined_posneg.values == 'X'))]
mylist = list(dict.fromkeys(mylist))
# mylist
```

```python
# Creates a new column filled with 'Y' then fills each row with the respective source (pos or neg).
# The ones that remains as Y will be latter replaced with 'both'

joined_posneg['Source'] = 'Y'
for item in mylist:
    if item[1] == 'pos':
        joined_posneg['Source'][item[0]] = 'neg'
    else:
        joined_posneg['Source'][item[0]] = 'pos'

joined_posneg['Source'].head()
```

```python
# Here we replace 'X' with 'NaN', and 'Y' with the string 'both' indicating that the rows that were not filled in the last step
# are present in both joined dataframes
joined_posneg.replace('X', np.nan, inplace=True)
joined_posneg.replace('Y', 'both', inplace=True)
joined_posneg.to_csv('joined_posneg.tsv', sep='\t')
joined_posneg
```

```python
# removing the column "Source"
joined_posneg = joined_posneg.drop(columns='Source')

# Rename columns, removing '_pos' and '_neg' suffixes
joined_posneg.columns = [i.split('_')[0] for i in joined_posneg.columns]

# group data based on the columns names, summing the values where columns have the same name
joined_posneg = joined_posneg.groupby(level=0, axis=1).sum()

# Replace values greater than 1 with ones
joined_posneg[joined_posneg>1] = 1

# fill NaN's with zero (NaN's appear where columns have the same name, but one of them is empty for some compound)
# and save to csv
joined_posneg.fillna(0).to_csv('joined_single_bin.tsv', sep='\t')
joined_posneg
```
